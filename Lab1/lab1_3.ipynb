```python
#Pandas (створити dataframe різними способами, провести маніпуляції з даними (head, describe, iloc, loc ...))

import pandas as pd

#Create
df = pd.DataFrame({
     'Film': ['DC studio', 'Dune', 'James Bond', 'Marvel','American History'],
     'Rating': [6.0, 5.8, 7.7, 8.1, 4.9],
     'Budget': [1.4, 2.2, 3, 11, 0.8]
}, index=['DC', 'Du', 'JB', 'MaR','AH'])
df.index.name = 'Filt studio code'
print("\nThe data frame is\n",df)
#Head
print("\nHead of 2\n",df.head(2))
#Describe
print("\nDescribe\n",df.describe())
#loc
print("\nLoc from 'Du' to 'AH' is\n",df.loc['Du':'AH', :])
#Iloc
print("\nLoc of '3' is\n",df.iloc[0])
#Sort
print("\nSort by Rating > 6.5\n",df[df.Rating > 6.5][['Film', 'Budget']])
#Add Series
df['Top Actor'] = [4, 5, 4, 20, 2]
print("\nAdd one series\n",df)
#Drop

#df.drop(['Top Actor'], axis='columns') Не работает почему то =(
#print("\nDrop one series\n",df)
del df['Top Actor']
print("\nDrop one series\n",df)
```
